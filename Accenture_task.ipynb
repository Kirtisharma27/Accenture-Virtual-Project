```python
import pandas as pd
df1 = pd.read_csv('C:\\Users\\Kirti Sharma\\Downloads\\Reactions (1).csv')
```

```python
df2 = pd.read_csv('C:\\Users\\Kirti Sharma\\Downloads\\Content (1).csv')
```

```python
df3 = pd.merge(df1,df2,on='Content ID', how='left')
df3
```

```python
k = df3.drop(['User ID_y','Unnamed: 0_x','Unnamed: 0_y','URL'], axis = 'columns')
k
```

```python
k.columns[k.isna().any()]
```

```python
k_1 = k.dropna(subset = ['Type_x'])
k_1
```

```python
df4 = pd.read_csv('C:\\Users\\Kirti Sharma\\Downloads\\ReactionTypes (1).csv')
```

```python
merge = pd.merge(k_1,df4,left_on='Type_x',right_on='Type',how='left')
```

```python
merge['Category']=merge['Category'].str.lower()
merge.drop(['Unnamed: 0'],axis='columns',inplace=True)
merge
```

```python
merge.drop(['Type_x','Sentiment'],axis='columns',inplace=True)
merge
```

```python
merge['Category'] = merge['Category'].str.replace('"','')
```

```python
m_group = merge.groupby('Category')
```

```python
sco = m_group.sum()
sco
```

```python
sco.sort_values(['Score'],ascending=False,inplace=True)
print(sco)
sco_ = sco.head(5)
print(sco_)
sco_.plot(kind='pie',autopct='%0.2f%%',subplots=True)
```

```python
with pd.ExcelWriter('C:\\Users\\Kirti Sharma\\Downloads\\Acc_kirti.xlsx') as writer:
    merge.to_excel(writer,sheet_name='data_without_grouping',index=False)
    sco.to_excel(writer,sheet_name='data_after_groupbycategory')
    sco_.to_excel(writer,sheet_name='top5_categories')
```
